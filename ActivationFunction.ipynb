```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linear Activation Function

def linear(x):
    return x

x = np.linspace(1,10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function : Linear')
plt.show()
```

```python
# Sigmoid Activation Function
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-10,10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function : Sigmoid')
plt.show()
```

```python
# Tanh Activation Function

def tanh(x):
    return np.tanh(x)

x = np.linspace(-10,10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function : Tanh')
plt.show()
```

```python
# Relu Activation Function

def ReLu(x):
    return np.maximum(0,x)
            
x = np.linspace(-10,10)
plt.plot(x, ReLu(x))
plt.axis('tight')
plt.title('Activation Function : ReLu')
plt.show()
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)

x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function : Softmax')
plt.show()
```
